```python
#Verificamos que no haya un id_producto repetido como ean y como NO ean
mask = productos_df.groupby('id_producto')['productos_ean'].nunique() > 1

# Filtrar el DataFrame original con los valores que cumplen la condición
df_filtrado = productos_df[productos_df['id_producto'].isin(mask[mask].index)]
df_filtrado.head()
```

```python
columnas_de_interes = [ 
       'id_sucursal',
       'id_producto',
       'productos_ean',
       'productos_descripcion',
       'productos_cantidad_presentacion',
       'productos_unidad_medida_presentacion',
       'productos_marca',
       'productos_precio_lista',
       'productos_precio_referencia',
       'productos_cantidad_referencia',
       'productos_unidad_medida_referencia',
       'productos_precio_unitario_promo1',
       'productos_leyenda_promo1',
       'productos_precio_unitario_promo2',
       'productos_leyenda_promo2'
]
```

```python
productos_df = productos_df[columnas_de_interes]
productos_df.head()
```

```python
productos_df['id_sucursal'] = productos_df['id_sucursal'].astype(int)
productos_df['productos_ean'] = productos_df['productos_ean'].astype(int).astype(bool)
productos_df['id_producto'] = productos_df['id_producto'].astype(int).astype(str)
productos_df['productos_unidad_medida_presentacion'] = productos_df['productos_unidad_medida_presentacion'].astype('category')
productos_df['productos_marca'] = productos_df['productos_marca'].astype('category')
productos_df['productos_unidad_medida_referencia'] = productos_df['productos_unidad_medida_referencia'].astype('category')
productos_df['precio_zscore'] = productos_df.groupby('id_producto')['productos_precio_referencia'].transform(
    lambda x: (x - x.mean()) / x.std()
)
productos_df = productos_df[productos_df['productos_descripcion'].str.contains('yerba', case=False, na=False)]
productos_df.head()
```

De la ultima celda podemos analizar que hay productos que estan en TODAS las sucursales y que hay productos que no

```python
# Leer el archivo Excel
sucursales_df = pd.read_excel('../sucursales/sucursales_procesado.xlsx')
sucursales_df.info()
```

```python
#Uno los dataframe por la columna de id_sucursal
df_completo = pd.merge(productos_df, sucursales_df, on='id_sucursal', how='left')
df_completo.head()
```

```python
# Vamos a analizar las sucursales que son del tipo supermercado
completo_supermercado_df = df_completo[
    (df_completo['sucursales_provincia'] == 'Buenos Aires') | 
    (df_completo['sucursales_provincia'] == 'CABA')
]
completo_supermercado_df = completo_supermercado_df[completo_supermercado_df['sucursales_tipo'] == 'Supermercado']
completo_supermercado_df.head()
```

```python
# Contar el número de sucursales únicas por producto
productos_en_todas_sucursales = completo_supermercado_df.groupby('id_producto')['comercio_bandera_nombre'].nunique()

# Filtrar los productos que están en todas los comercios, que en nuestro caso son 3, expres, hiper, market
productos_en_todas_sucursales = productos_en_todas_sucursales[productos_en_todas_sucursales == 3].index

# Filtrar el DataFrame original
productos_df_en_varias_sucursales = completo_supermercado_df[completo_supermercado_df['id_producto'].isin(productos_en_todas_sucursales)]
print(f"Cantidad de productos que estan en los 3 comercios {len(productos_df_en_varias_sucursales['id_producto'].value_counts())}")
productos_df_en_varias_sucursales["id_producto"].value_counts()
```

```python
# Agrupar por id_producto y comercio_bandera_nombre y calcular el máximo y mínimo de precio_zscore
resultado = productos_df_en_varias_sucursales.groupby(['id_producto', 'comercio_bandera_nombre'])['precio_zscore'].agg(['max', 'min']).reset_index()

resultado["id_producto"].value_counts()
```

```python
resultado.head()
```

```python
# Ordenar el DataFrame por la columna 'id_producto'
resultado = resultado.sort_values(by='max').reset_index(drop=True)

# Convertir la columna 'id_producto' a categórica
resultado['id_producto'] = resultado['id_producto'].astype('category')

# Agregar una nueva columna con el código numérico de la categoría
resultado['id_producto_code'] = resultado['id_producto'].cat.codes

resultado.head(20)
```

```python
plt.figure(figsize=(12, 6))
df_sorted = resultado.sort_values(by=['comercio_bandera_nombre', 'id_producto_code'])

sns.lineplot(
    data=df_sorted,
    x='id_producto_code',
    y='min',
    hue='comercio_bandera_nombre',       # Diferenciar por tipo
    palette='Set1',   # Paleta de colores (opcional)
    estimator=None,      # Mostrar todos los puntos (sin estimación)
    legend='full'           # Tamaño de los puntos
)

# Agregar título y etiquetas
plt.title('Precio vs Código por Tipo')
plt.xlabel('Código')
plt.ylabel('Precio')
plt.legend(title='Tipo')  # Opcional: título de la leyenda
plt.show()
```
