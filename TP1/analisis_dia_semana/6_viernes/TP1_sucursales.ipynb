```python
import pandas as pd
import zipfile
```

```python
# Ruta al archivo zip
zip_path = 'viernes.zip'

# Abrir el zip
with zipfile.ZipFile(zip_path) as z:
    # Abrir solo el archivo CSV deseado
    with z.open('sucursales.csv') as f:
        sucursales_df = pd.read_csv(f, sep='|')

sucursales_df.head()
```

```python
#notamos que el ultimo registro de cada csv es irrelevante
sucursales_df = sucursales_df.drop(sucursales_df.index[-1])
```

```python
columnas_de_interes = [ 'id_bandera', 
                        'id_sucursal', 
                        'sucursales_nombre',
                        'sucursales_tipo', 
                        'sucursales_latitud', 
                        'sucursales_longitud',
                        'sucursales_barrio', 
                        'sucursales_localidad',
                        'sucursales_provincia']
```

```python
sucursales_df = sucursales_df[columnas_de_interes]
sucursales_df.head()
```

```python
# Leer el archivo Excel
provincias_df = pd.read_excel('../../sucursales/maestro-provincias.xlsx')

# Mostrar las primeras filas
provincias_df.head(25)
```

```python
sucursales_df['sucursales_provincia'] = sucursales_df['sucursales_provincia'].map(provincias_df.set_index('sucursales_provincia')['provincia'])
sucursales_df.head()
```

```python
sucursales_df['id_bandera'] = sucursales_df['id_bandera'].astype(int)
sucursales_df['id_sucursal'] = sucursales_df['id_sucursal'].astype(int).astype(str)
sucursales_df.head()
```

```python
sucursales_df.to_excel('sucursales_procesado.xlsx', index=False)
```
