Determinamos las columnas de interés.

```python
sucursales_df.columns
```

```python
columnas_de_interes = ['id_bandera', 
                       'id_sucursal', 
                       'sucursales_nombre',
                       'sucursales_tipo', 
                       'sucursales_latitud', 
                       'sucursales_longitud',
                       'sucursales_barrio', 
                       'sucursales_localidad',
                       'sucursales_provincia']
```

```python
sucursales_df = sucursales_df[columnas_de_interes]
sucursales_df.head()
```

Reemplazamos el código de provincia por el nombre según el maestro de provincias.

```python
# Leer el archivo Excel
provincias_df = pd.read_excel('../../maestro-provincias.xlsx')

# Mostrar las primeras filas
provincias_df.head(25)
```

```python
sucursales_df['sucursales_provincia'] = sucursales_df['sucursales_provincia'].map(provincias_df.set_index('sucursales_provincia')['provincia'])
sucursales_df.head()
```

```python
sucursales_df['id_bandera'] = sucursales_df['id_bandera'].astype(int)
sucursales_df['id_sucursal'] = sucursales_df['id_sucursal'].astype(int).astype(str)
sucursales_df.head()
```

Mapeamos el id_bandera al nombre del tipo de comercio.

```python
# Leer el archivo Excel
comercio_df = pd.read_excel('../comercio/comercio_procesado.xlsx')
comercio_df.info()
```

```python
sucursales_df['comercio_bandera_nombre'] = sucursales_df['id_bandera'].map(comercio_df.set_index('id_bandera')['comercio_bandera_nombre'])
```

```python
plt.hist(sucursales_df['comercio_bandera_nombre'])

plt.show()
```

Analizamos las sucursales por provincia.

```python
sucursales_cantidad = sucursales_df['sucursales_provincia'].value_counts()

fig, ax = plt.subplots(1, 2, figsize=(15, 8))

sucursales_cantidad.plot(kind='barh',  ax=ax[0])
ax[0].set_title('Cantidad de sucursales por provincias')
ax[0].set_ylabel('Provincia')

sucursales_cantidad.plot(kind='barh',  ax=ax[1])
ax[1].set_title('Cantidad de sucursales por provincias (con zoom)')
ax[1].set_ylabel('Provincia')
ax[1].set_xlim(0, 50)

plt.tight_layout()  # Ajusta automáticamente los subgráficos para que no se solapen

plt.show()
```

Analizamos las sucursales por provincia y tipo de comercio.

```python
plt.figure(figsize=(15, 8))
cross_tab = pd.crosstab(sucursales_df['sucursales_provincia'], sucursales_df['comercio_bandera_nombre'])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Número de sucursales por provincia y comercio')
plt.xlabel('Comercio')
plt.ylabel('Provincia')
plt.show()
```

Analizamos las sucursales por provincia y tipo de sucursal.

```python
plt.figure(figsize=(15, 8))
cross_tab = pd.crosstab(sucursales_df['sucursales_provincia'], sucursales_df['sucursales_tipo'])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Número de sucursales por provincia y tipo de sucursal')
plt.xlabel('Tipo de sucursal')
plt.ylabel('Provincia')
plt.show()
```

Analizamos las sucursales por tipo de sucursal y tipo de comercio.

```python
plt.figure(figsize=(15, 8))
cross_tab = pd.crosstab(sucursales_df['comercio_bandera_nombre'], sucursales_df['sucursales_tipo'])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Número de sucursales tipo de sucursal y comercio')
plt.xlabel('Tipo de sucursal')
plt.ylabel('Comercio')

plt.show()
```

Con esto podemos concluir que no hay una relacion directa entre el tipo de comercio y el tipo de sucursal.

Además, dado que en los cross_tab previos de tipo de sucursal por provincia se puede observar que la gran mayoría de las sucursales se encuentran en la Ciudad Autónoma de Buenos Aires y en la provincia de Buenos Aires nos quedamos solo con las sucursales de estas jurisdicciones.

```python
sucursales_df = sucursales_df[
    (sucursales_df['sucursales_provincia'] == 'Buenos Aires') | 
    (sucursales_df['sucursales_provincia'] == 'CABA')
]
```

Volvemos a generar el cross_tab luego de este filtro.

```python
plt.figure(figsize=(15, 8))
cross_tab = pd.crosstab(sucursales_df['comercio_bandera_nombre'], sucursales_df['sucursales_tipo'])
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Número de sucursales tipo de sucursal y comercio')
plt.xlabel('Tipo de sucursal')
plt.ylabel('Comercio')

plt.show()
```

Viendo este último cross_tab podemos notar que el tipo de sucursal que está presente para cada tipo de comercio es Supermercado, pero como para el tipo de comercio Hipermercado Carrefour hay solo una sucursal de este tipo decidimos tener en cuenta además el tipo de sucursal Hipermercado para no sesgar los datos.

```python
sucursales_df = sucursales_df[sucursales_df['sucursales_tipo'].isin(['Hipermercado', 'Supermercado'])]
```

```python
sucursales_df["localidad_provincia"] = sucursales_df["sucursales_localidad"] + " - " + sucursales_df["sucursales_provincia"]
```
